```python
# days = 365
# weeks =52
# months =12

age = int(input('What is your age? '))

years_remaining = 90-age
days_remaining = years_remaining*365
weeks_remaining = years_remaining*52
months_remaining = years_remaining*12

print(f'You got {years_remaining} years, {months_remaining} months, {weeks_remaining} weeks and {days_remaining} days to live.')
```

## Tip Calculator

```python
print('Welcome to the tip calculator')
bill = float(input('What was the bill? $'))
pct = float(input('What percentage tip would you like to give? '))
people = int(input('How many people to split the bill'))


tip = float((bill*pct)/100)

each_person_pay = round((bill + tip)/people,2)

print(f'Each person should pay : {each_person_pay} $')
```
